```python
import json
import math

# Load the JSON file
json_path = "3dfullres_fall_summary.json"  # Update this path if required
with open(json_path, "r") as file:
    data = json.load(file)

# Print mean Dice scores for each class
print("Mean Dice Scores for Each Class:")
for class_id, metrics in data["mean"].items():
    print(f"  Class {class_id}: Dice Score = {metrics['Dice']:.4f}")

# Dice score threshold
THRESHOLD = 0.2

# Perform patient-level classification for each case
print("\nPatient-Level Classification Results:")
for case in data["metric_per_case"]:
    metrics = case["metrics"]
    case_name = case["prediction_file"].split("/")[-1]  # Extract filename

    # Get Dice scores for each class
    dice_1 = metrics.get("1", {}).get("Dice", 0)
    dice_2 = metrics.get("2", {}).get("Dice", 0)
    dice_3 = metrics.get("3", {}).get("Dice", 0)
    dice_4 = metrics.get("4", {}).get("Dice", 0)
    
    print(f"{dice_1}, {dice_2}, {dice_3}, {dice_4}")

    # Apply classification rules
    if dice_1 >= THRESHOLD:
        patient_label = "TAAD"
    elif dice_2 >= THRESHOLD and (dice_1 < THRESHOLD or math.isnan(dice_1)):
        patient_label = "TBAD"
    else:
        patient_label = "NoAD"

    #print(f"  {case_name}: {patient_label}")
```
